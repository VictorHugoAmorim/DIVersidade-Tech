```python
def atracoes_liberadas(ano_de_nascimento, altura):
    atracoes = []
    altura_minima_carrossel = 1
    idade_minima_carrossel = 3
    idade_minima_piscina_de_bolinhas = 4
    idade_maxima_piscina_de_bolinhas = 9
    altura_maxima_piscina_de_bolinhas = 1.30
    altura_minima_montanha_russa = 1.10
    idade = 2022 - ano_de_nascimento
    
    
    if altura_minima_carrossel <= altura and idade_minima_carrossel <= idade:
        atracoes.append('Carrossel Liberado ')
    else:
        pass
    if idade_minima_piscina_de_bolinhas <= idade and idade_maxima_piscina_de_bolinhas >= idade and altura_maxima_piscina_de_bolinhas >= 1.30:
        atracoes.append('Piscina de bolinhas Liberada ')
    else:
        pass
    if altura_minima_montanha_russa <= altura:
        atracoes.append('Montanha russa Liberada ')
    else:
        pass
    return [x for x in atracoes]
            
atracoes_liberadas(2018,1.30)
```

```python
num = int(input("Digite um número: "))
contador = 0
divisores = []

for i in range(1, num + 1):
    if num % i == 0:
        divisores.append(i)
        contador += 1
print(f'Os divisores são {divisores}')

if contador == 2:
    print("O número é primo")
else:
    print("O número não é primo")
```

```python
numero_anterior = 0
numero_repetido = list()
for n in range(10):
    numero_digitado = int(input(f'Digite o {n+1}º número: \n'))
    if numero_digitado == numero_anterior:
        numero_anterior = numero_digitado
        numero_repetido.append(numero_digitado)
    else:
        pass
    numero_anterior = numero_digitado

if len(numero_repetido) > 0:
    print(f'O número {max(numero_repetido)} foi repetido em sequencia')
else:
    print('Nenhum numero foi repetido em sequencia')

```
