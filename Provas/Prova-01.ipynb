```python
numero = int(input('Digite o númmero:\n'))
lista_de_numeros = []

for n in range(numero):
    lista_de_numeros.append(n+1)
lista_de_numeros.reverse()
for n in range(len(lista_de_numeros)):
    print(lista_de_numeros[len(lista_de_numeros)-(n+1):])
```

```python
lista_de_notas = []
while True:
    nota = float(input('Digite uma nota:\nPara finalizar o programa, digite "-1"\n'))
    if nota >=0:
        lista_de_notas.append(nota)
    elif nota < -1:
        print('Por favor, digite um nº maior ou igual a 0')
    else:
        break
        
        
        
media = sum(lista_de_notas) / len(lista_de_notas)
maior = max(lista_de_notas)
menor = min(lista_de_notas)

print('='*40)
print(f'''
Média de notas: {media}
Maior nota: {maior}
MKenor notas: {menor}
        ''')
```

```python
def comparador_de_texto(txt1,txt2):
    texto_1 = txt1
    texto_2 = txt2

    if texto_1.upper() in texto_2.upper():
        print(True)
    else:
        print(False)
        
comparador_de_texto('olá','Olá, João! Tudo bem?')
```

```python
def numeros_inteiros_pares(inteiros):
    lista_pares = []
    if len(inteiros) > 10:
        print('favor, digitar até 10 números')
    else:
        pass
    for n in range(len(inteiros)):
        if inteiros[n] % 2 == 0:
            lista_pares.append(inteiros[n])
        else:
            pass
    return lista_pares

numeros_inteiros_pares([1,2,3,4,5,6,7,8,9])
```
