```python
nota1 = float(input('Digita a 1º nota:\n'))
nota2 = float(input('Digita a 2º nota:\n'))
media = (nota1 + nota2) /2

if media == 10:
    print('Aprovado com Distinção')
elif media >= 7:
    print('Aprovado')
else:
    print('Reprovado')
```

```python
num1 = int(input('1º nº: \n'))
num2 = int(input('2º nº: \n'))
num3 = int(input('3º nº: \n'))
maior = num1

if (num1 > num2) and (num1>num3):
    maior=num1
elif (num2 > num3) and (num2>num1):
    maior=num2
else:
    maior=num3

print(f'O maior nº é {maior}')
```

```python
turno = input('Qual turno vc estuda?\n(M)Maturino\n(V)Vespertino\n(N)Noturno')

while (turno.upper() not in 'MVB'):
    print('\nValor inválido\n')
    turno = input('Qual turno vc estuda?\n(M)Maturino\n(V)Vespertino\n(N)Noturno')
if turno.upper() == 'M':
    print('Bom dia!')
elif turno.upper() == 'V':
    print('Boa tarde!')
else:
    print('Boa noite!')
```

```python
salario = float(input('Digite seu salário:\n'))
novo_salario = 0

if salario <= 280:
    novo_salario += salario * 1.2
elif (salario > 280) and (salario < 700):
    novo_salario += salario * 1.15
elif (salario > 700) and (salario < 1500):
    novo_salario += salario * 1.1
else:
    novo_salario += salario * 1.05

print(f'Salario antes do reajuste: R${salario}')
print(f'Percentual do aumento: {round(novo_salario/salario-1,2)}%')
print(f'Total do aumento: R${round(novo_salario - salario)}')
print(f'Novo salario após reajuste: R${novo_salario}')
```
