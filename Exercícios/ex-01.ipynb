```python
renda_mensal = float(input('Renda mensal:\n'))
emprestimo = float(input('emprestimo:\n'))
numero_de_prestacoes = int(input('Prestações :\n'))

if (emprestimo <= (renda_mensal *10)):
    numero_de_prestacoes = emprestimo / numero_de_prestacoes
    if numero_de_prestacoes > (renda_mensal * 0.3):
        print('Emprestimo não concedido pois a parcela excede 30%')
    else:
        print('Emprestimo concedido')
else:
    print('Emprestimo não concedido pois o valor é maior que a renda *10')
```
